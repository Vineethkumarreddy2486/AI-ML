```python
# 1. deleting rows where empty cells exists ::  dropna()

df = pd.read_csv("../files/practice_file.csv")
print("before cleaning the empty cells")
print(df.to_string())
# df[df.isnull().any(axis=1)]             # to check where empty cells exactly


print("removing the rows where the empty cells exists in any column")
cleaned_df = df.dropna()
print(cleaned_df.to_string())

print("removing only the rows where ' All time peak ' values are empty")

cleaned_peak_df = df.dropna(subset=['All Time Peak'])
print(cleaned_peak_df.to_string())

```

```python
# filling rows with some random where empty cells exists ::  fillna()

df = pd.read_csv("../files/practice_file.csv")
print("before filling the values")
print(df.to_string())


print("after filling the values")
fill_df = df.fillna({"All Time Peak":10,"Actual gross":20})
print(fill_df.to_string())
```

```python
# filling rows mean(), median(), mode() where empty cells exists ::  fillna()

df = pd.read_csv("../files/practice_file.csv")
print("before filling the values")
print(df.to_string())


print("after filling the values")
all_time_peak_mean = df["All Time Peak"].mean()
peak_mean = df["Peak"].mean()

all_time_peak_median = df["All Time Peak"].median()
peak_median = df["Peak"].median()

all_time_peak_mode = df["All Time Peak"].mode()
peak_mode = df["Peak"].mode()

print("the mean of the 'All Time Peak column' : ", all_time_peak_mean)
print("the mean of the 'Peak column' : ", peak_mean)

print("the median of the 'All Time Peak column' : ", all_time_peak_median)
print("the median of the 'Peak column' : ", peak_median)

print("the mode of the 'All Time Peak column' : ", all_time_peak_mode)
print("the mode of the 'Peak column' : ", peak_mode)

# filling all time peak empty values with all time peak mean and actual gross with peak mean

print(df.fillna({'All Time Peak':all_time_peak_mean,'Actual gross':peak_mean}))

```
